```python
from flask import  Flask , request, jsonify
import pymongo

client = pymongo.MongoClient("mongodb+srv://arshewin:mongodb12@cluster0.wckow.mongodb.net/?retryWrites=true&w=majority")
data_base = client['pandas_task']
collection = data_base["fitbit"]

app = Flask(__name__)

@app.route('/insert',  methods =['GET' , 'POST'])
def mangodb_insert():
    if(request.method=='POST'):
        a = request.json['Insert']
        result = collection.insert_one(a)
        return jsonify((str(result)))

@app.route('/update',  methods =['GET' , 'POST'])
def mangodb_update():
    if(request.method=='POST'):
        a = request.json['query']
        b = request.json['new_value']
        result = collection.update_one(a,b)
        return jsonify((str(result)))



@app.route('/delete',  methods =['GET' , 'POST'])
def mangodb_delete():
    if(request.method=='POST'):
        c = request.json['query_del']
        result = collection.delete_one(c)
        return jsonify((str(result)))


@app.route('/find',  methods =['GET' , 'POST'])
def mangodb_find():
    if(request.method=='POST'):
        d = request.json['query']
        result1 = collection.find_one(d)
        return jsonify((str(result1)))

if __name__=='__main__'  :
    app.run()

```
